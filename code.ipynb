```python
def CreateChild(Matrix,row):
    LowestCost=[]
    Matrixes=[]
    for col in range(8):
        for i in range(8):
            Matrix[row][i]='O'  
        Matrix[row][col]='Q'
        LowestCost.append(CaculateHeuristic(Matrix))
    Min=min(LowestCost)
    cols = [i for i, v in enumerate(LowestCost) if v == Min]
    print(cols)
    for j in cols:
        for i in range(8):
            Matrix[row][i]='O'  
        Matrix[row][j]='Q'
        Matrixes.append(Matrix)
        ls=[Matrixes,Min]
    return ls
```

```python
def CaculateRows(Matrix):
    Sum=0
    for i in range(8):
        Queen=0
        for j in range(8):
            if Matrix[i][j]=='Q':
                Queen=Queen+1
        if Queen>=1:
            Sum=Sum+((Queen*(Queen-1))//2)
    return Sum
```

```python
def CaculateCols(Matrix):
    Sum=0
    for j in range(8):
        Queen=0
        for i in range(8):
            if Matrix[i][j]=='Q':
                Queen=Queen+1
        if Queen>=1:
            Sum=Sum+((Queen*(Queen-1))//2)
    return Sum
```

```python
def CaculateDiagnoleRightRows(Matrix):
    Sum=0
    for col in range(8):
        Queen=0
        j=col
        for i in range(8):
            if j>7 :
                break
            if Matrix[i][j]=='Q':
                Queen=Queen+1
            j=j+1
        if Queen>=1:
            Sum=Sum+((Queen*(Queen-1))//2)
    return Sum
```

```python
def CaculateDiagnoleLeftRows(Matrix):
    Sum=0
    for col in range(7,-1,-1):
        Queen=0
        j=col
        for i in range(8):
            if j<0:
                break
            if Matrix[i][j]=='Q':
                Queen=Queen+1
            j=j-1
        if Queen>=1:
            Sum=Sum+((Queen*(Queen-1))//2)
    return Sum
```

```python
def CaculateDiagnoleRightCols(Matrix):
    Sum=0
    for row  in range(1,8,1):
        Queen=0
        i=row
        for j in range(8):
            if i>7 :
                break
            if Matrix[i][j]=='Q':
                Queen=Queen+1
            i=i+1
        if Queen>=1:
            Sum=Sum+((Queen*(Queen-1))//2)
    return Sum
    
```

```python
def CaculateDiagnoleLeftCols(Matrix):
    Sum=0
    for row  in range(1,8,1):
        Queen=0
        i=row
        for j in range(7,-1,-1):
            if i>7 :
                break
            if Matrix[i][j]=='Q':
                Queen=Queen+1
            i=i+1
        if Queen>=1:
            Sum=Sum+((Queen*(Queen-1))//2)
    return Sum
```

```python
def CaculateHeuristic(Matrix):
    SumAll=0
    SumRows=CaculateRows(Matrix)
    SumCols=CaculateCols(Matrix)
    SumDiagnoleRightRows=CaculateDiagnoleRightRows(Matrix)
    SumDiagnoleLefttRows=CaculateDiagnoleLeftRows(Matrix)
    SumDiagnoleRightCols=CaculateDiagnoleRightCols(Matrix)
    SumDiagnoleLeftCols=CaculateDiagnoleLeftCols(Matrix)

    SumAll=SumRows+SumCols+SumDiagnoleRightRows+SumDiagnoleLefttRows+SumDiagnoleRightCols+SumDiagnoleLeftCols
    return SumAll
        
```
